```python
#training and testing
from sklearn.model_selection import train_test_split
x_train,x_test=train_test_split(data.drop(columns='price'))
y_train,y_test=train_test_split(data['price'])
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
predict=reg.predict(x_test)
print("Linear Regression:")
reg.score(x_train,y_train)
```

```python
#Random forest regressor
from sklearn.ensemble import RandomForestRegressor
reg1=RandomForestRegressor()
reg1.fit(x_train,y_train)
predict1=reg.predict(x_test)
print("Random Forest")
reg1.score(x_train,y_train)
```

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor
reg2=DecisionTreeRegressor()
reg2.fit(x_train,y_train)
predict2=reg2.predict(x_test)
reg2.score(x_train,y_train)
```

```python
#mean square and mean absolute error
from sklearn.metrics import mean_absolute_error,mean_squared_error
prediction=[predict,predict1,predict2]
algorithms=["LinearRegression","RandomForest","DecisionTree"]
c=0
for i in prediction:
        print(algorithms[c], "\n", "MeanSquaredError-",mean_squared_error(y_test,i),"\n","MeanAbsoluteError-",mean_absolute_error(y_test,i))
        c+=1

        
```
